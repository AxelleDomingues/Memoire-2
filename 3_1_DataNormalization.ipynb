```python
from sklearn import preprocessing
import pandas as pd
artp = pd.read_excel("CLEANnormMIDIAFR.xlsx")
scaler = preprocessing.MinMaxScaler()
names = artp.columns
d = scaler.fit_transform(artp)
scaled_df = pd.DataFrame(d, columns=names)
scaled_df.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("CLEANResultNormMIDIAFRtoUse.xlsx")
sns.set(rc={'figure.figsize':(25,10), 'grid.color':'0.6', 'grid.linestyle': ":", 'axes.facecolor':'white', 'figure.facecolor':'lightgrey', "figure.dpi":72, 'savefig.dpi':72})

a = sns.scatterplot(data=df, x="Période", y="Normalisation", edgecolor="black", hue="Corpus", ci=None, palette=["plum", "khaki"], size="Occurrences", alpha=0.7, sizes=(2000, 20000), legend="full")
h,l = a.get_legend_handles_labels()
plt.legend(h[1:3],l[1:3],bbox_to_anchor=(0, 1), loc=2, borderaxespad=1, shadow=True, fancybox=True,  title ="Légende", fontsize = '20', title_fontsize = "20")

a.tick_params(bottom=True, left=True)
a.set(xticklabels = ([' ', '1200-1375', '1376-1532', '1533-1691', '1692-1840']))

plt.xlabel('Période', fontsize=20, fontweight='bold')
plt.ylabel('Normalisation', fontsize=20, fontweight='bold')
plt.xticks((np.arange(0,5,1)), fontsize=15)
plt.yticks((np.arange(-0.1,1.2,0.1)), fontsize=15)
plt.xlim(0,5)
plt.ylim(-0.1,1.2)
```
