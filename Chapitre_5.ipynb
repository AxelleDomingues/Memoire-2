```python
dfFRa = pd.read_excel ('PrepFusAFrVariation.xlsx')
dfFRde = pd.read_excel ('PrepFusDeFrVariation.xlsx')

dfFRal=dfFRa[(dfFRa['Forme']=="al")]
dfFRau=dfFRa[(dfFRa['Forme']=="au") | (dfFRa['Forme']=="aus") | (dfFRa['Forme']=="aux") | (dfFRa['Forme']=="auz") | (dfFRa['Forme']=="av") | (dfFRa['Forme']=="AVX")]
dfFRaul=dfFRa[(dfFRa['Forme']=="aulx")]

dfFRdu=dfFRde[(dfFRde['Forme']=="du") | (dfFRde['Forme']=="DV")]
dfFRdel=dfFRde[(dfFRde['Forme']=="del") | (dfFRde['Forme']=="dels")]
dfFRdes=dfFRde[(dfFRde['Forme']=="de") | (dfFRde['Forme']=="des") | (dfFRde['Forme']=="dês") | (dfFRde['Forme']=="dez")]
dfFRdudit=dfFRde[(dfFRde['Forme']=="dudit")]
dfFRdesdits=dfFRde[(dfFRde['Forme']=="desdits") | (dfFRde['Forme']=="desdites")]



sns.set(rc={'figure.figsize':(30,25), 'grid.color':'black', 'axes.facecolor':'white', 'figure.facecolor':'lightgrey', "figure.dpi":72, 'savefig.dpi':72})
fig, ax = plt.subplots(nrows=2, ncols=1, sharey=True)
fig.text(0.07, 0.5, "Densité", va='center', rotation='vertical', fontsize='30', fontweight='bold')

plt.subplot(2,1,1)
sns.kdeplot(dfFRal[' '], color='forestgreen', linewidth="4")
sns.kdeplot(dfFRau[' '], color='lime', linewidth="4")
sns.kdeplot(dfFRaul[' '], color='yellow', linewidth="4")
plt.legend(labels=["Al", "Au, aus, aux, auz", "Aulx"], borderpad=1, shadow=True, fancybox=True, ncol=1, title ="Légende", fontsize = '18', title_fontsize = "18", loc = 2, bbox_to_anchor = (0,1))
#plt.xlabel('Date', fontsize=15)
#plt.ylabel('Densité', fontsize=15)
plt.xticks((np.arange(1200,1840,50)), fontsize=15)
plt.yticks((np.arange(0,0.012,0.002)), fontsize=15)
plt.xlim(1200,1840)
plt.ylim(0,0.012)


plt.subplot(2,1,2)
sns.kdeplot(dfFRdu['Date'], color='blue', linewidth="4")
sns.kdeplot(dfFRdel['Date'], color='dodgerblue', linewidth="4")
sns.kdeplot(dfFRdes['Date'], color='cyan', linewidth="4")
sns.kdeplot(dfFRdudit['Date'], color='purple', linewidth="4")
sns.kdeplot(dfFRdesdits['Date'], color='magenta', linewidth="4")
plt.legend(labels=["du","del", "des, dez, dês", "dudit", "desdits"], borderpad=1, shadow=True, fancybox=True, ncol=1, title ="Légende", fontsize = '18', title_fontsize = "18", loc = 2, bbox_to_anchor = (0,1))
plt.xlabel('Date', fontsize='30', fontweight='bold')
#plt.ylabel('Densité', fontsize=15)
plt.xticks((np.arange(1200,1840,50)), fontsize=15)
plt.yticks((np.arange(0,0.012,0.002)), fontsize=15)
plt.xlim(1200,1840)
plt.ylim(0,0.012)
```
