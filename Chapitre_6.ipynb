```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("Partitif.xlsx")
sns.set(rc={'figure.figsize':(35,30), 'grid.color':'0.6', 'grid.linestyle': ":", 'axes.facecolor':'white', 'figure.facecolor':'lightgrey', "figure.dpi":72, 'savefig.dpi':72})
fig, ax = plt.subplots(nrows=3, ncols=1, sharex=True, sharey=True)
fig.text(0.07, 0.5, "FTR sur 1 million d'occurences", va='center', rotation='vertical', fontsize='30', fontweight='bold')

plt.subplot(2,1,1)
a = sns.scatterplot(data=df, x="  ", y="   ", linewidth=4, edgecolor="white", hue="Corpus", palette=["crimson", "orange"], size="   ", sizes=(3000, 30000), alpha=0.85, legend="full")
h,l = a.get_legend_handles_labels()
plt.legend(h[1:3],l[1:3],bbox_to_anchor=(0, 1), loc=2, borderaxespad=1, shadow=True, fancybox=True,  title ="Légende", fontsize = '25', title_fontsize = "25")

a.tick_params(bottom=True, left=True)
a.set(xticklabels = ([' ', '1200-1375', '1376-1532', '1533-1691', '1692-1840']))

#plt.xlabel('Période', fontsize=25, fontweight='bold')
#plt.ylabel("FTR sur 1M d'occurrences", fontsize=25, fontweight='bold')
plt.xticks((np.arange(0,5,1)), fontsize=20)
plt.yticks((np.arange(0,40000,5000)), fontsize=20)
plt.xlim(0,5)
plt.ylim(0,40000)
yticklabel=False

plt.subplot(2,1,2)
a = sns.scatterplot(data=df, x="  ", y="   ", linewidth=4, edgecolor="white", hue="Corpus", palette=["crimson", "orange"], size="Occurrences", sizes=(3000, 30000), alpha=0.85, legend="full")
h,l = a.get_legend_handles_labels()
plt.legend(h[1:3],l[1:3],bbox_to_anchor=(0, 1), loc=2, borderaxespad=1, shadow=True, fancybox=True,  title ="Légende", fontsize = '25', title_fontsize = "25")

a.tick_params(bottom=True, left=True)
a.set(xticklabels = ([' ', '1200-1375', '1376-1532', '1533-1691', '1692-1840']))

plt.xlabel('Période', fontsize=30, fontweight='bold')
#plt.ylabel("FTR sur 1M d'occurrences", fontsize=25, fontweight='bold')
plt.xticks((np.arange(0,5,1)), fontsize=20)
plt.yticks((np.arange(0,40000,5000)), fontsize=20)
plt.xlim(0,5)
plt.ylim(0,40000)
yticklabel=False
```
